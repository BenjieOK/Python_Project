```python
import csv
import numpy as np
import pandas as pd
import datetime

with open('../data/AAPL-2016.csv', newline='') as csv_file:
    data = csv.DictReader(csv_file)
    apple_stock = np.empty((0,2))
    for row in data:
        myArray = np.array([[row['Date'], row['Volume']]])
        apple_stock = np.append(apple_stock, myArray, axis=0)

date = [datetime.datetime.strptime(x,'%Y-%m-%d') for x in apple_stock[:,0]]
dat = pd.Series(np.float64(apple_stock[:,1]), index=date)
dat.plot()
```
